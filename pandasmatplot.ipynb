## Python matplotlib pandas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Creating a simple NumPy array
# Let's create a range of numbers from 0 to 10 with an interval of 0.1
x = np.arange(0, 10, 0.1)
print("Value of x is   ",x)

# Step 2: Perform some mathematical operations using NumPy
# Let's calculate the sine of each element in the array
y = np.sin(x)
print("value of y is ",y)

# Step 3: Plotting the data using Matplotlib
plt.plot(x, y, label='sin(x)')  # plot x against sin(x)  ###### Main plot function
plt.title("Plot of sine function")  # Add a title
plt.xlabel("x values")  # Label for x-axis
plt.ylabel("y values")  # Label for y-axis
plt.legend()  # Display legend
plt.grid(True)  # Display grid

# Step 4: Show the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create data using NumPy
# Let's create an array of x values from 0 to 10 with a step size of 1
x = np.arange(0, 11, 1)  # This will generate values [0, 1, 2, ..., 10]
print("Value of x is ",x)

# Step 2: Define a simple linear relationship for y values
# For example, y = 2x + 1
y = 2 * np.sin(x) + 1
print("Value of y is ",y)

# Step 3: Create a line chart using Matplotlib
plt.plot(x, y, label="y = 2x + 1")  # Plot x and y values    ####MAIn function

# Step 4: Adding title and labels
plt.title("Simple Line Chart")  # Title of the plot
plt.xlabel("X values")  # X-axis label
plt.ylabel("Y values")  # Y-axis label

# Step 5: Display legend
plt.legend()

# Step 6: Display the grid
plt.grid(True)
plt.savefig("line_chart.png")# Step 7: Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create data using NumPy
# Let's create an array of categories (for the x-axis)
categories = ['A', 'B', 'C', 'D', 'E']   # Main point

# Create an array of values (for the y-axis)
values = np.array([3, 7, 2, 5, 4])   # Main point

# Step 2: Create a bar plot using Matplotlib
plt.bar(categories, values, color='skyblue')  ###MAIN point Create a bar plot with categories on x-axis and values on y-axis

# Step 3: Adding title and labels
plt.title("Simple Bar Plot")  # Title of the plot
plt.xlabel("Categories")  # X-axis label
plt.ylabel("Values")  # Y-axis label

# Step 4: Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create random data using NumPy
# Generate 1000 random numbers from a normal distribution (mean=0, std=1)
data = np.random.randn(1000)

# Step 2: Create a histogram plot using Matplotlib
plt.hist(data, bins=30, color='skyblue', edgecolor='black')  # 30 bins, color, and edge color

# Step 3: Adding title and labels
plt.title("Histogram of Random Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Step 4: Show the plot
plt.show()
```
